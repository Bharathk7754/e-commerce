```python
# prompt: read the file /content/Bharath K - revolutioncart_data.csv

import pandas as pd
df = pd.read_csv('/content/Bharath K - revolutioncart_data.csv')
df
```

```python
# prompt: split the data into x and y  where y is monthly_revenue and x is rest

y = df['monthly_revenue']
x = df.drop('monthly_revenue', axis=1)
```

```python
# prompt: split the x and y into train and test

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# prompt:  make ensemble model considering  regression ,
# SVM, knn and DT and provide the model accuracy for x train and y train

from sklearn.ensemble import VotingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Create individual regression models
lr_model = LinearRegression()
svm_model = SVR()
knn_model = KNeighborsRegressor()
dt_model = DecisionTreeRegressor()

# Create an ensemble model using VotingRegressor
ensemble_model = VotingRegressor(estimators=[
    ('lr', lr_model),
    ('svm', svm_model),
    ('knn', knn_model),
    ('dt', dt_model)
])

# Train the ensemble model
ensemble_model.fit(x_train, y_train)

# Make predictions on the training data
y_train_pred = ensemble_model.predict(x_train)

# Evaluate the model's performance on the training data
mse_train = mean_squared_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

print(f"Training MSE: {mse_train}")
print(f"Training R-squared: {r2_train}")
```
